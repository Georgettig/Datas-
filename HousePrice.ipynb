```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

# Base Train 

Importando a base de dados train

```python
train = pd.read_csv('train.csv')
train
```

Verificando porcentagem de dados nulos por coluna

```python
(train.isnull().sum()/train.shape[0]).sort_values(ascending=False).head(20)
```

Deletando colunas que possuem mais de 10% de dados nulos

```python
deletar = train.columns[(train.isnull().sum()/train.shape[0]) > 0.1]
deletar
```

```python
train = train.drop(deletar, axis=1)
```

Eliminando dados categóricos da base de dados

```python
texto = train.columns[train.dtypes != 'object']
texto
```

```python
train = train.loc[:,texto]
train
```

Verificando a existência de dados nulos

```python
train.isnull().sum().sort_values(ascending=False).head()
```

Preenchendo dados nulos de cada coluna com suas respectivas médias

```python
media_garage = round(train['GarageYrBlt'].mean(),1)
media_garage
```

```python
media_area = round(train['MasVnrArea'].mean(),1)
media_area
```

```python
train['GarageYrBlt'] = train['GarageYrBlt'].fillna(media_garage)
```

```python
train['MasVnrArea'] = train['MasVnrArea'].fillna(media_area)
```

```python
train
```

Importando biblioteca para criação das bases de treino e teste

```python
from sklearn.model_selection import train_test_split
```

```python
X = train.drop('SalePrice', axis=1)
X
```

```python
y = train.SalePrice
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

Regressão Linear

```python
from sklearn.linear_model import LinearRegression
```

```python
linear = LinearRegression().fit(X_train, y_train)
```

```python
previsao_linear = linear.predict(X_test)
```

Árvore de Decisão

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree = DecisionTreeRegressor().fit(X_train, y_train)
```

```python
previsao_tree = tree.predict(X_test)
```

SVM

```python
from sklearn.svm import SVC
```

```python
svm = SVC().fit(X_train, y_train)
```

```python
previsao_svm = svm.predict(X_test)
```

Erro médio absoluto e Erro quadrárico médio

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
print(f'Erro Absoluto Regressão Linear: {mean_absolute_error(y_test, previsao_linear)}')
print(f'Erro Quadrático Regressão Linear: {mean_squared_error(y_test, previsao_linear)}')
```

```python
print(f'Erro Absoluto Árvore: {mean_absolute_error(y_test, previsao_tree)}')
print(f'Erro Quadrático Árvore: {mean_squared_error(y_test, previsao_tree)}')
```

```python
print(f'Erro Absoluto SVM: {mean_absolute_error(y_test, previsao_svm)}')
print(f'Erro Quadrático SVM: {mean_squared_error(y_test, previsao_svm)}')
```
